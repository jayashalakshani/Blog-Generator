```python
import os
import subprocess
import threading
from langchain import PromptTemplate
from langchain.llms import CTransformers
import streamlit as st
from pyngrok import ngrok
```

```python
# Install necessary packages
!pip install streamlit pyngrok langchain
```

```python
# Enviromnet secret keys
from google.colab import userdata
sec_key = userdata.get('HF_TOKEN')
# print(sec_key)
```

```python
# # Set Hugging Face token securely
os.environ['HUGGINGFACEHUB_API_TOKEN'] = sec_key
```

```python
# Hugging Face Llama-2 model repository ID
repo_id = "TheBloke/Llama-2-7B-Chat-GGML"
```

```python
# Function to generate a response using the model
def getLLamaresponse(input_text, no_words, blog_style):
    llm = CTransformers(model=repo_id, model_type='llama', config={'max_new_tokens': 256, 'temperature': 0.01})

    # Prompt template for generating blog content
    template = """
        Write a blog for {blog_style} job profile on the topic '{input_text}' within {no_words} words.
    """

    # Create a prompt template
    prompt = PromptTemplate(template=template, input_variables=['input_text', 'no_words', 'blog_style'])

    # Format the prompt using the provided inputs
    formatted_prompt = prompt.format(input_text=input_text, no_words=no_words, blog_style=blog_style)

    # Generate response
    response = llm(formatted_prompt)
    return response
```

```python
# Start the Streamlit app in a separate thread
def run_streamlit():
    subprocess.run(["streamlit", "run", "app.py", "--server.port=8501", "--server.headless=true"])
```

```python
# Write Streamlit app code to a file
with open("app.py", "w") as f:
    f.write("""\
import os
import streamlit as st
from langchain import PromptTemplate
from langchain.llms import CTransformers

# Set Hugging Face token securely
os.environ['HUGGINGFACEHUB_API_TOKEN'] = 'HUGGINGFACEHUB_API_TOKEN'
repo_id = "TheBloke/Llama-2-7B-Chat-GGML"

def getLLamaresponse(input_text, no_words, blog_style):
    llm = CTransformers(model=repo_id, model_type='llama', config={'max_new_tokens': 256, 'temperature': 0.01})

    template = \"\"\"
        Write a blog for {blog_style} job profile on the topic '{input_text}' within {no_words} words.
    \"\"\"

    prompt = PromptTemplate(template=template, input_variables=['input_text', 'no_words', 'blog_style'])
    formatted_prompt = prompt.format(input_text=input_text, no_words=no_words, blog_style=blog_style)

    response = llm(formatted_prompt)
    return response

st.set_page_config(page_title="Llama 2 Chatbot", page_icon=":robot:", layout="centered", initial_sidebar_state="collapsed")
st.header("Generate Blogs ")
input_text = st.text_input("Enter the Blog Title")

# Additional input columns
col1, col2 = st.columns([5, 5])
with col1:
    no_words = st.text_input("No of words")
with col2:
    blog_style = st.selectbox("Writing the blog for", ("Researchers", "Data Scientist", "Common People","Student","Software Engineer"), index=0)

submit = st.button("Generate")

# Generate blog when the button is pressed
if submit:
    if input_text and no_words.isdigit():
        response = getLLamaresponse(input_text, no_words, blog_style)
        st.write(response)
    else:
        st.write("Please enter a valid blog title and word count.")
    """)
```

```python
# Start the Streamlit app
streamlit_thread = threading.Thread(target=run_streamlit)
streamlit_thread.start()
```

```python
# ngrok integration to expose the Streamlit app
ngrok.set_auth_token("ngrok_token")
public_url = ngrok.connect(8501)
print(f"Streamlit app is running at: {public_url}")
```
